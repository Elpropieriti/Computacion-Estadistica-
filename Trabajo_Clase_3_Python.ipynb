1)Convertir "de","dl" a un sistema de coordendas polares 


```python
import pandas as pd
import numpy as np
#Creacion funcion de crecimiento
def prog(r, n, a1):
  an = a1 + r*(n-1)
  seq = np.arange(start=a1, stop=an, step=r)
  return seq

prog(r=7, n=20, a1=15)

np.random.seed(123)
#DataFrame de los diametros
df1 = pd.DataFrame({
    'de': np.sort(np.random.normal(loc = 4, scale = 1, size=96)),
    'dl': np.sort(np.random.normal(loc=4.5, scale=1.2, size=96)),
    'ddd': np.repeat(prog(r=7, n=25, a1=15), 4)
})

df1['localidad'] = np.repeat(['l1','l2']*24, 2)
df1.head()
#Conversion a coordendas polares corresponde a los diametros de(x),dl(y))
df1.assign(r=np.sqrt((df1["de"]/2)**2+(df1["de"]/2)**2),
           theta=np.arctan2(df1["de"],df1["de"]))
```

2) Graficar un Cardioide


```python
import numpy as np
import math
import matplotlib.pyplot as plt

#Para crear un Cardiode en funcion de Factor A y B.
a=float(input("Ingrese un valor A:"))
b=float(input("Ingrese un valor B:"))

theta = np.linspace(0, 2*np.pi, 1000)

r = a + b * np.cos(theta)

plt.polar(theta, r, 'r')

plt.show()
```
