$y_{2}=log(\sqrt {x})$

1)$Condicion:x>0$


```python
x = float(input('Ingrese x: '))
#importamos la libreria "math" de Python
import math
#usar funciones Srt(raices) y Funcion Log(logaritmos)
if x >0:
  resultado=math.sqrt(x)
  resultado=math.log(resultado)
  print(f'El resultado es {resultado}')
else:
    print(' Ingrese un valor mayor a 0')
```

$y_{3}=log(\sqrt {x^{2}-1})$
1)$Condicion: x<-1$ $o$ $x>1$ 


```python
x = float(input('Ingrese x: '))
#importamos la libreria "math" de Python
import math
#usar funciones Srt(raices) y Funcion Log(logaritmos)
if -1<=x<=1:
  print(' Ingrese un valor  fuera del intervalo de [-1 , 1]')
else:
  x=x**2-1
  resultado=math.sqrt(x)
  resultado=math.log(resultado)
  print(f'El resultado es {resultado}')
```

$y_{4}=\frac{log(\sqrt {x -1})}{x^2+1}$

1)$Condicion: $x>1$ 

```python
x = float(input('Ingrese x: '))
#importamos la libreria "math" de Python
import math
#usar funciones Srt(raices) y Funcion Log(logaritmos)
if x<1:
  print(' Ingrese un valor  mayor a 1')
else:
  denominador=x**2+1
  numerador=math.sqrt(x-1)
  numerador=math.log(numerador)
  resultado=numerador / denominador
  print(f'El resultado es {resultado}')
```

$y_{5}=\frac{x-1}{log(x)}$

1)$Condicion:(0,1) U (1,∞)$

```python
x = float(input('Ingrese x: '))
#importamos la libreria "math" de Python
import math
#usar funciones Srt(raices) y Funcion Log(logaritmos)
if x==1 or x<0:
  print(' Ingrese un dentro el intervalo (0,1) U (1,∞)')
else:
  denominador=x-1
  numerador=math.log(x)
  resultado=numerador / denominador
  print(f'El resultado es {resultado}')
```

$y_{6}=\frac{1}{x}+\sqrt {x -1}$

1)$Condicion:x>1 $

```python
x = float(input('Ingrese x: '))
#importamos la libreria "math" de Python
import math
#usar funciones Srt(raices) y Funcion Log(logaritmos)
if x<1:
  print(' Ingrese un valor mayor a 1')
else:
  log=math.sqrt(x-1)
  resultado=1/x+log
  print(f'El resultado es {resultado}')
```

$y_{7}=\frac{1}{x}+\sqrt {x -1}-\frac{1}{log(\frac{1}{x+1})}$

1)$Condicion:x>1 $

```python
x = float(input('Ingrese x: '))
#importamos la libreria "math" de Python
import math
#usar funciones Srt(raices) y Funcion Log(logaritmos)
if x<1:
  print(' Ingrese un valor mayor a 1')
else:
  log=math.log(1/(x+1)
  sqr=math.sqrt(x-1)
  resultado=1/x+sqr-log
  print(f'El resultado es {resultado}')
```
