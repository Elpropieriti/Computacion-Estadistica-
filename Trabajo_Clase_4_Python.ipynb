1)Autocorrelation of Time Series Data in Python

```python
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
##Leer la serie

data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Serie caudales diarios.csv")
nan=data.isnull().sum()
print(nan)#Datos faltantes o nulos

data = data[['Fecha', 'Serie Qobs']].set_index(['Fecha'])
data = data.dropna()# eliminar datos "NaN"

##Graficar

plt.rcParams["figure.figsize"] = (20,3)
plt.plot(data)
plt.title(label='Serie de tiempo del caudal')
#plt.title("Linear graph")
plt.xlabel('AÃ±o')
plt.ylabel('m^3/s')
plt.show()

```

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt
##Autocorrelacion
plot_acf(data)
# Show the AR as a plot
plt.show()
```

3. Calcular CV (Coeficiente de variaciÃ³n)


```python
import pandas as pd
import numpy as np

#Crear un dataframe sintetico

rows=17
cols=4
dat=pd.DataFrame(np.random.randint(0, 100, (rows, cols)), columns=['A', 'B', 'C', 'D'])## Las claves columns deben tener deben tener la misma cantidad que cols
print(dat)

#Crear la funcion Lambda con las fucniones mean(promedio) y std(desviacion standar) librerias Panda y Numpy

cv=lambda x:np.std(x)/np.mean(x)

cv(dat)
```


4. Seleccionar un tamaÃ±o de muestra

$n=\frac{N\sigma^{2} Z^{2}}{(N-1)e^{2}+\sigma^{2} Z^{2}}$


```python
import pandas as pd
import numpy as np
#Funcion Lambda
n_e=lambda sig,n,z,e:(n*(sig**2)*(z**2))/((n-1)*e**2)+((sig**2)*(z**2))
vlue=n_e(0.5,9000,1.96,0.05)
print(f"El valor minimo de muestra esperado es: {vlue}")
```


5. Convetir absorbancia en transmitancia


```python
import pandas as pd
import numpy as np

#Creacion de una lista de valores de absorbancia.

ab = np.linspace(0.30, 0.99 , 90)
print(ab)
#Creacion de la funcion.
tr=lambda x: 10**-x
r_tr=tr(ab)
print(r_tr)
```
