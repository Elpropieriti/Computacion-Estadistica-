```python
import pandas as pd
import pingouin as pg
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import seaborn as sns
```

```python
df= pd.read_excel('/content/drive/MyDrive/Colab Notebooks/datos_CE_1001083581.xlsx')
df.head()
```

```python
pg.anova(data = df, dv='cbd', between='accesion')
```

```python
formula = ols('cbd ~ C(accesion)', data=df).fit()
anova_lm(formula)
```

```python
pg.normality(formula.resid, method='shapiro')
```
